```python
def get_article_year(url):
    """Функция для получения года публикации статьи с страницы"""
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        # Находим год публикации, обычно он находится в мета-тегах
        year = None
        for meta in soup.find_all('meta'):
            if 'name' in meta.attrs and meta.attrs['name'] == 'citation_year':
                year = meta.attrs.get('content')
                break
        return year if year else "Не указан"
    except Exception as e:
        print(f"Ошибка при извлечении года для {url}: {e}")
        return "Не указан"

def search_google_scholar(query, num_results=10, year=None):
    """Функция для поиска статей на Google Scholar с фильтрацией по году"""
    search_url = f"https://scholar.google.com/scholar?q={query.replace(' ', '+')}"
    
    # Если год задан, добавляем параметры для фильтрации по году
    if year:
        search_url += f"&as_ylo={year}&as_yhi={year}"  # Фильтруем статьи по указанному году
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    response = requests.get(search_url, headers=headers)
    
    # Проверка на успешность запроса
    if response.status_code != 200:
        print(f"Ошибка при запросе: {response.status_code}")
        return []
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Находим все элементы с ссылками на статьи
    articles = []
    
    for result in soup.find_all('h3', {'class': 'gs_rt'}):
        link = result.find('a')
        if link:
            title = link.text
            url = link['href']
            year = get_article_year(url)  # Получаем год публикации статьи
            articles.append({'title': title, 'url': url, 'year': year})
        
        # Ограничиваем количество результатов
        if len(articles) >= num_results:
            break
    
    return articles

def save_to_csv(articles, filename="aviander.csv"):
    """Сохранение результатов в CSV файл"""
    # Преобразуем список словарей в DataFrame
    df = pd.DataFrame(articles)
    
    # Удаляем дубликаты по URL
    df = df.drop_duplicates(subset='url', keep='first')
    
    # Сохраняем в CSV
    df.to_csv(filename, index=False)
    print(f"Сохранено в {filename}")

# Пример использования
query = "Авиандр"  # Здесь указываешь название препарата или ключевое слово
year = None  # Укажи нужный год (или можно оставить None для всех годов)
results = search_google_scholar(query, num_results=10, year=year)
print(results)
save_to_csv(results)
```

```python

def search_google_scholar(query, num_results, year=None, max_pages=5):
    """Функция для поиска статей на Google Scholar с фильтрацией по году и пагинацией"""
    articles = []
    
    for page in range(max_pages):
        # Формируем URL для поиска с пагинацией
        start = page * 10  # Параметр start для пагинации (каждая страница содержит 10 результатов)
        search_url = f"https://scholar.google.com/scholar?q={query.replace(' ', '+')}&start={start}"
        
        # Если год задан, добавляем параметры для фильтрации по году
        if year:
            search_url += f"&as_ylo={year}&as_yhi={year}"
        
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        
        response = requests.get(search_url, headers=headers)
        
        # Проверка на успешность запроса
        if response.status_code != 200:
            print(f"Ошибка при запросе: {response.status_code}")
            break
        
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Находим все элементы с ссылками на статьи
        for result in soup.find_all('h3', {'class': 'gs_rt'}):
            link = result.find('a')
            if link:
                title = link.text
                url = link['href']
                articles.append({'title': title, 'url': url})
        
        # Ограничиваем количество результатов
        if len(articles) >= num_results:
            break
        
        # Задержка, чтобы не перегрузить сервер
        time.sleep(2)
    
    return articles

def save_to_csv(articles, filename="exenatide.csv"):
    """Сохранение результатов в CSV файл"""
    # Преобразуем список словарей в DataFrame
    df = pd.DataFrame(articles)
    
    # Удаляем дубликаты по URL
    df = df.drop_duplicates(subset='url', keep='first')
    
    # Сохраняем в CSV
    df.to_csv(filename, index=False)
    print(f"Сохранено в {filename}")

# Пример использования
query = "Exenatide"  # Здесь указываешь название препарата или ключевое слово
year = 2024  # Укажи нужный год (или можно оставить None для всех годов)
results = search_google_scholar(query, num_results=50, year=year, max_pages=5)
print(results)
save_to_csv(results)
```

```python

def search_google_scholar(query, year, max_results=1000):
    """Функция для поиска статей на Google Scholar за определенный год с пагинацией"""
    articles = []
    page = 0
    while len(articles) < max_results:
        # Формируем URL для поиска с пагинацией
        start = page * 10  # Параметр start для пагинации (каждая страница содержит 10 результатов)
        search_url = f"https://scholar.google.com/scholar?q={query.replace(' ', '+')}&as_ylo={year}&as_yhi={year}&start={start}"
        
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        
        response = requests.get(search_url, headers=headers)
        
        # Проверка на успешность запроса
        if response.status_code != 200:
            print(f"Ошибка при запросе: {response.status_code}")
            break
        
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Находим все элементы с ссылками на статьи
        new_articles = []
        for result in soup.find_all('h3', {'class': 'gs_rt'}):
            link = result.find('a')
            if link:
                title = link.text
                url = link['href']
                new_articles.append({'title': title, 'url': url, 'year': year})
        
        # Если не было новых статей, завершаем поиск
        if not new_articles:
            print(f"Не найдено новых статей на странице {page + 1}. Завершаем сбор.")
            break
        
        # Добавляем новые статьи в общий список
        articles.extend(new_articles)
        
        # Задержка, чтобы избежать блокировки со стороны Google
        time.sleep(2)
        
        # Переход к следующей странице
        page += 1
    
    return articles

def save_to_csv(articles, filename="rifaximin.csv"):
    """Сохранение результатов в CSV файл"""
    # Преобразуем список словарей в DataFrame
    df = pd.DataFrame(articles)
    
    # Удаляем дубликаты по URL
    df = df.drop_duplicates(subset='url', keep='first')
    
    # Сохраняем в CSV
    df.to_csv(filename, index=False)
    print(f"Сохранено в {filename}")

# Пример использования
query = "Rifaximin"  # Здесь указываешь название препарата или ключевое слово
year = 2024  # Указываем нужный год для поиска
results = search_google_scholar(query, year, max_results=1000)  # Максимум 1000 статей
print(f"Найдено {len(results)} статей.")
save_to_csv(results)
```

```python
# с фильтрацией по открытому доступу
def search_google_scholar(query, num_results=10):
    # Формируем URL для поиска
    search_url = f"https://scholar.google.com/scholar?q={query.replace(' ', '+')}"
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    response = requests.get(search_url, headers=headers)
    
    # Проверка на успешность запроса
    if response.status_code != 200:
        print(f"Ошибка при запросе: {response.status_code}")
        return []
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Находим все элементы с ссылками на статьи
    articles = []
    
    for result in soup.find_all('h3', {'class': 'gs_rt'}):
        link = result.find('a')
        if link:
            title = link.text
            url = link['href']
            
            # Проверка на наличие ссылки на PDF (открытый доступ)
            pdf_link = result.find_next('div', {'class': 'gs_or_ggsm'})
            if pdf_link:
                # Это значит, что у статьи есть открытый доступ (например, PDF)
                open_access = True
            else:
                open_access = False
            
            if open_access:
                articles.append({'title': title, 'url': url})
        
        # Ограничиваем количество результатов
        if len(articles) >= num_results:
            break
    
    return articles

def save_to_csv(articles, filename="scholar_articles.csv"):
    # Преобразуем список словарей в DataFrame
    df = pd.DataFrame(articles)
    
    # Сохраняем в CSV
    df.to_csv(filename, index=False)
    print(f"Сохранено в {filename}")

# Пример использования
query = "Fabomotizole"  # Укажите ключевое слово или название препарата
results = search_google_scholar(query, num_results=10)
print(results)
save_to_csv(results)
```
