Функция map позволяет применить какую-то функцию к каждому элементу списка.

```python
l = ['1', '2', '3']
# каждый строковый элемент из l теперь стал числовым
list(map(int, l))
```

zip нужен для создания нового списка из "пар" соответствующих элементов.

```python
n1 = (1, 2, 3)
n2 = ('one', 'two', 'three')
list(zip(n1, n2))
```

```python
# можно еще добавить и третий, и четвертый списки и тд
n3 = ('uno', 'dos', 'tres')
list(zip(n1, n2, n3))
```

Рекурсия - это использование функции внутри себя самой

```python
# например
def factorial (n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
students = pd.read_csv('/Users/olgazdanova/Downloads/students.csv')
students
```

```python
# сгруппируем оценки по группе и выведем минимальную для каждой группы
students.groupby('Group').aggregate(min)
```

```python
# а теперь выводится и минимальная, и максимальная оценка
students.groupby('Group').aggregate([min, max])
```

```python
# применим ту же функцию, но к конкретной колонке (Mark)
students.groupby('Group').aggregate({'Mark': [min, max]})
```

```python
# также можем написать свои собственные функции и применять их к столбцам
def grants_count(col):
    return col[col>3].count()
students.groupby('Group').aggregate({'Mark':[min, max, grants_count]})
# получается в группе А на стипендию претендует 2 человека, а в группе Б - 3
```

```python
# имена методов (а count в данном случае метод) обязательно передаем в кавычках, иначе не будет работать
students.groupby('Group').aggregate({'Mark':[min, max, grants_count, 'count']})
```

```python
students[['Group', 'Mark']].groupby(['Group']).mean()
```

```python
# созданную колонку можно переименовать 
students[['Group', 'Mark']].groupby('Group').mean().rename(columns={'Mark':'Средний балл'}) 
```

```python
new = students.groupby('Group').aggregate({'Mark':[min, max, 'mean', grants_count, 'count']})
new.rename(columns={'min':'Минимальный балл', 'max':'Максимальный балл', 'mean':'Средний балл', 'grants_count':'Получат стипендию', 'count':'Всего студентов'})
```
