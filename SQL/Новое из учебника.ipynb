Инструкция SELECT TOP используется для указания кол-ва возвращаемых записей. Актуально при работе с большими БД с тысячами записей. SELECT TOP поддерживается не везде - например, в MySQl ту же функцию выполняет LIMIT, а в Oracle - ROWNUM 

# Подстановочный знак (с LIKE)

в MS Access: * - ноль или более символов; ? - 1 символ; [] - любой отдельный символ в скобках; ! - любой символ, не заключенный в скобки; [-] - набор символов из перечня в скобках; # - любой числовой символ

в SQL Server: % - ноль или более символов; _ - 1 символ; [] - любой символ в скобках; [^] - любой символ не в скобках; [-] - набор символов из перечня в скобках

Например, задание. Выбрать все записи, в которых первая буква в столбце Город это "а", или "с", или "ц". 
решение: SELECT * FROM Customers WHERE City LIKE '[асц]%'
