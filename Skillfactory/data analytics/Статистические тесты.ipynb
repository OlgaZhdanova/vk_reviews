Параметр генеральной совокупности — это константа, конкретное число, которое, как правило, неизвестно и никогда не станет известным.

Благодаря статистическим тестам можно делать выводы о параметрах всей генеральной совокупности, имея только выборку наблюдений.

Без статистических тестов выводы, полученные на выборке, нельзя распространить на всю генеральную совокупность.

Статистическая гипотеза — это некоторое утверждение о параметрах генеральной совокупности, которое будет проверяться на основе данных выборки.

Формулировка гипотез должна быть математически корректной и ссылаться на параметры генеральной совокупности, а не выборки. Например, утверждение «рекомендательная система влияет на средний чек покупки» — это статистическая гипотеза.

Общий алгоритм проверки гипотезы состоит из шести этапов:

- Построение нулевой и альтернативной гипотез.

Нулевая гипотеза формулируется так, чтобы быть консервативной (утверждать отсутствие эффекта; статус-кво), а альтернативная — исследует эффект или изменение.
- Выбор уровня значимости .

Этот уровень показывает вероятность ошибочно отвергнуть нулевую гипотезу (ошибка первого рода).
- Сбор данных для проверки гипотезы.

Важно учитывать репрезентативности выборок и устранения систематических ошибок.
- Выбор статистического теста.

Тест выбирается с учетом поставленной задачи, характера данных и количества выборок.
- Проведение статистического теста, вычисление p-value.

p-value — это вероятность получить наблюдаемые данные (или еще более экстремальные значения), если нулевая гипотеза верна.

- Сравнение p-value c уровнем значимости  и вывод, отклонить или не отклонить нулевую гипотезу.
- Если p-value меньше , нулевая гипотеза отвергается в пользу альтернативной.
- Если p-value больше или равно , оснований для отклонения нулевой гипотезы недостаточно.

В ходе статистических тестов вы будете проверять нулевую гипотезу. Возможен один из двух результатов:

- отклоняем нулевую гипотезу в пользу альтернативной (делаем вывод, что эффект есть);
- не отклоняем нулевую гипотезу, так как для этого недостаточно оснований (нельзя сделать вывод о наличии эффекта).

Есть несколько типов альтернативных гипотез:

- Двусторонняя гипотеза (): тест обнаруживает любые изменения (увеличение или уменьшение).
- Односторонняя гипотеза ( или ): тест обнаруживает изменения в определенном направлении.
- Левостороннее исследование (): интересуют только случаи, где меньше .
- Правостороннее исследование (): интересуют только случаи, где больше .

Ошибки при проверке гипотез:

- Ошибка I (первого) рода — отклонение нулевой гипотезы, когда она на самом деле верна.

Эта ошибка также называется альфа-ошибкой, или ложноположительным результатом.

- Ошибка II (второго) рода — происходит, когда вы не отклоняете нулевую гипотезу, хотя она на самом деле ложна.

Эта ошибка также называется бета-ошибкой, или ложноотрицательным результатом.

- Вероятность допустить ошибку I рода обычно задается заранее (например, 5%, что соответствует уровню значимости р = 0.05).

- Мощность теста (1 - бета) — это вероятность обнаружить эффект, если он в действительности существует. Например, если мощность составляет 0.93, это значит, что вероятность ошибки II рода равна 0.07, или 7%.

!!! Важно понимать, что p-value (уровень статистической значимости) — это именно вероятность получить наблюдаемые или более экстремальные значения, если верна нулевая гипотеза, а не вероятность верности нулевой гипотезы.

Уровень значимости (альфа) — это заранее установленный порог вероятности, с которым допускается риск отклонения нулевой гипотезы, когда она на самом деле верна (ошибка I рода). Обычно принимают , что соответствует 5%-ному риску ошибочного отклонения Н0.


!!! Если полученное значение p-value меньше (либо равно) заранее заданного уровня значимости альфа, нулевая гипотеза отвергается.

Еще раз 

- Если полученное p-value ниже уровня значимости (альфа), результаты статистически значимы и согласуются с альтернативной гипотезой.
- Если полученное p-value выше уровня значимости (альфа), результаты считаются статистически незначимыми, нельзя отвергнуть нулевую гипотезу.

Универсального значения альфа не существует, так как уровень значимости зависит от приемлемого уровня риска, который может варьироваться в зависимости от ситуации. Обычно выбирают 0.01 (строгий порог; в медицине), 0.05 (самый распротсраненный; в маркетинге) или 0.10 (реже, где допустимы ошибки 1 рода).

Чем более нежелательно допустить ошибку I рода, тем меньшее значение  стоит выбирать. 

Если вы минимизируете альфа (риск ошибки I рода), делая статистический тест более консервативным, вы увеличиваете бета (риск ошибки II рода).

Обратное тоже верно: снижение бета делает тест менее строгим и увеличивает альфа.

Ключевые факторы, влияющие на выбор статистического теста:

- количество анализируемых групп (выборок) — может быть одна, две или более -> тесты мб одновыборочными, двухвыборочными и многовыборочными;
- зависимость между группами (могут быть зависимыми или независимыми) -> тесты мб парные и непарные, соответственно;
- знание о распределении признаков (важно понять, соблюдают ли данные нормальное распределение). Если данные распределены нормально, к ним можно применить параметрические тесты

Параметрические тесты — это статистические методы, основанные на параметрах нормального распределения (среднее значение и стандартное отклонение).

Если вы знаете, что признаки не распределены нормально или у нас нет информации об их распределении, необходимо использовать НЕпараметрические тесты. Их можно применять на любых количественных признаках независимо от вида их распределения. Непараметрические методы менее чувствительны к выбросам в данных.

Данные на нормальность распределения можно проверить с помощью графиков.

А можно с помощью статистических методов (более надежно). Один из таких примеров тест Шапиро-Уилка; хорошо работает на выборках до 5 тысяч объектов.

Гипотезы в основе теста:

- нулевая — данные распределены нормально.
- альтернативная — данные не распределены нормально.

При интерпретации результата теста Шапиро-Уилка, если p-value теста больше уровня значимости альфа, можно считать, что данные соответствуют нормальному распределению.

```python
import scipy.stats as stats
# проверим кол-во заказов в день на нормальность
# задаём уровень значимости
alpha = 0.05 

# загружаем данные
data = [31, 38, 30, 58, 40, 36, 38, 44, 43, 39, 36, 41, 37, 20, 38, 35, 38, 11, 42, 34]

# проводим тест Шапиро — Уилка
_, p = stats.shapiro(data)

print('p-value = %.3f' % (p))

# интерпретируем результат
if p <= alpha:
    print('Распределение не нормальное')
else:
    print('Распределение нормальное')
```

Чем примечателен Q-Q (квантиль-квантиль) plot? Он позволяет кроме принадлежности к нормальному распределению:
- оценить степень отклонения данных от теоретического распределения
- графически проиллюстрировать такие параметры как расположение данных, масштаб и скошенность. Читаем: медиану, дисперсию и наклон функции плотности распределения.
- сравнить две выборки между собой
- делать выводы, не основываясь на таких спорных показателях как p.value.

### Параметрические тесты

Одновыборочный t-критерий (t-test) используют, чтобы сравнить среднее значение признака с его заданным значением. Этот метод позволяет определить, существует ли статистически значимое отклонение среднего выборочного значения от известного (популяционного) значения.

Примеры задач, когда может понадобиться использование теста:

Необходимо проверить, что среднее время, которое пользователи проводят на сайте, больше пяти минут.
Необходимо установить, соответствуют ли химические показатели товара заявленным на упаковке.

```python
# задаём уровень значимости
alpha = 0.05 

# загружаем данные
data = [41, 38, 40, 46, 40, 46, 41, 44, 43, 39, 36, 41, 37, 45, 38, 45, 38, 48, 42, 34]

# проводим тест
# среди аргументов вводим выборочные данные, ожидаемое среднее, вид альтернативной гипотезы (в данном случае правосторонняя)
_, p = stats.ttest_1samp(data, popmean=40, alternative='greater')

print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```

Двухвыборочный t-критерий (t-test) используют, чтобы сравнить средние значения в двух независимых группах. Он помогает определить, имеются ли статистически значимые различия между средними значениями этих групп.

Двухвыборочный t-критерий может быть применен в следующих случаях:

- когда необходимо оценить эффект определенного воздействия на две группы;
- при сравнении новых методов, продуктов или процессов, чтобы проверить, есть ли значимая разница между ними.

!!! Ограничения применения теста:

- Группы должны быть независимыми.
- Перед использованием необходимо установить равенство (или неравенство) дисперсий в выборках. От этого зависит значение параметра equal_var.

```python
import numpy as np

# задаём уровень значимости
alpha = 0.05 

# загружаем данные (две независимые выборки)
data_zaryad = np.array([41,38,40,46,40,46,41,44,43,39,36,41,37,45,38,45,38,48,42,34])
data_planeta = np.array([40,39,42,46,41,46,42,45,44,42,38,42,38,46,39,46,40,41,43,36])

# проводим тест Левена на равенство дисперсий
print('Тест на равенство дисперсий')
result = stats.levene(data_planeta, data_zaryad)
p = result[1]
print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('Дисперсии не одинаковы, в stats.ttest_ind нужно использовать параметр equal_var=False.')
else:
    print('Дисперсии одинаковы, в stats.ttest_ind нужно использовать параметр equal_var=True.')

# проводим тест на сравнение средних в группах
print('\nТест на равенство средних')
_, p = stats.ttest_ind(data_planeta, data_zaryad, alternative='greater', equal_var=True)

print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```

Также на основе результатов теста Левена и t-теста можно сказать, что:

если p-value теста на равенство дисперсий больше уровня значимости, то считаем, что дисперсии равны, и используем `equal_var=True` в `stats.ttest_ind()`.
p-value теста t позволяет нам решить, можно ли отвергнуть нулевую гипотезу о равенстве средних значений.

!!! порядок выборок в параметрах stats.ttest_ind важен: он должен соответствовать альтернативной гипотезе.

Однофакторный дисперсионный анализ (ANOVA) используют, чтобы сравнить средние значения в более чем двух независимых группах. Допустим, есть ли статистически значимое влияние одного фактора (например, типа батарей) на зависимую переменную (например, время работы батареек).

Ограничения применения теста:

- Группы должны быть независимыми.
- Тест можно выполнять на группах с равной дисперсией.
- Тест можно применять, только если данные в выборках распределены нормально.


```python
# задаём уровень значимости
alpha = 0.05

# загружаем данные
group_zaryad = [31, 38, 30, 46, 40, 36, 38, 44, 43, 39, 36, 41, 37, 35, 38, 35, 38, 38, 42, 34]
group_planeta = [36, 45, 41, 41, 35, 32, 34, 42, 48, 43, 41, 39, 35, 34, 52, 42, 44, 43, 35, 43]
group_energiya = [35, 37, 39, 49, 45, 26, 46, 32, 49, 41, 48, 41, 47, 37, 45, 41, 43, 38, 40, 43]

# проводим тест Левена на равенство дисперсий
print('Тест на равенство дисперсий')
result = stats.levene(group_zaryad, group_planeta, group_energiya)
p = result[1]
print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('Дисперсии не одинаковы, нужно использовать непараметрический тест')
else:
    print('Дисперсии одинаковы, мы можем использовать тест ANOVA')

# проводим тест на сравнение средних в группах
print('\nТест на равенство средних')
_, p = stats.f_oneway(group_zaryad, group_planeta, group_energiya)

print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```

Парный t-критерий (t-test) используют, чтобы сравнить средние значения в двух ЗАВИСИМЫХ группах. Это важный инструмент в статистическом анализе, позволяющий проверять эффективность изменений в различных областях, таких как медицина, маркетинг и финансы. Для нормальных данных

```python
# задаём уровень значимости
alpha = 0.05 

# загружаем данные
data_zaryad = np.array([41, 34, 35, 47, 39, 42, 36, 43, 48, 38, 36, 47, 39, 32, 45, 40, 36, 39, 42, 46])
data_planeta = np.array([45, 42, 40, 43, 44, 41, 43, 39, 45, 45, 40, 43, 41, 42, 41, 41, 43, 46, 45, 42])

# проводим тест
_, p = stats.ttest_rel(data_planeta, data_zaryad, alternative='greater')
print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```

Однофакторный дисперсионный анализ (ANOVA) с повторными измерениями используют, чтобы сравнить средние значения в БОЛЕЕ ЧЕМ ДВУХ зависимых группах. Для нормальных данных

```python
from statsmodels.stats.anova import AnovaRM
import pandas as pd

# задаём уровень значимости
alpha = 0.05 

# создаём датасет
data = pd.DataFrame({'group': np.repeat(['Заряд', 'Планета', 'Энергия'], 5),
                          'object_id': np.tile([1, 2, 3, 4, 5], 3),                          
                          'value': [36, 45, 41, 41, 35, 32, 34, 42, 48, 43, 41, 39, 35, 34, 52]})


# проводим тест
print(AnovaRM(data=data, depvar='value', subject='object_id', within=['group']).fit())
```

### Непараметрические тесты

Такие тесты используются, когда данные не распределены нормально или об их распределении ничего не известно.

Особенности непараметрических методов заключаются в том, что они сравнивают не сами значения признаков, а их ранги или частоты.

В непараметрических тестах нулевые гипотезы будут о равенстве распределений признаков (что включает в себя равенство средних) или о равенстве медиан признаков

Типы непараметрических тестов:

- тест Уилкоксона — для зависимых выборок;
- критерий Манна-Уитни — для независимых выборок;
- критерий Краскала-Уоллиса — для сравнения более двух групп;
- тест знаков.

Как сформулировать нулевую и альтернативную гипотезы для теста, если у вас всего одна выборка?

В этом случае имеющаяся выборка (точнее, ее статистика, вычисляемая тестом) будет сравниваться со значением, заданным в параметрах теста:

- Нулевая гипотеза — соответствие (равенство) выборки и заданного значения.
- Альтернативная гипотеза — их отличие.

Критерий знаков (для одной группы) - непараметрический аналог одновыборочного t-критерия - используется, если нужно сравнить значения медианы признака с заданным значением.

```python
from statsmodels.stats.descriptivestats import sign_test

# задаём уровень значимости
alpha = 0.05 

# загружаем данные
data = [55, 53, 60, 49, 45, 57, 46, 53, 59, 53, 53, 55, 42, 41, 59, 43, 47, 60, 50, 57, 59, 56, 52, 48, 59, 53, 59, 50, 59, 59]

# проводим тест 
# выборочные данные и ожидаемое значение медианы
_, p = sign_test(data, 60)

print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```

U-критерий Манна-Уитни. Его используют, чтобы оценить различия распределений признака в двух независимых группах. Это непараметрический аналог двухвыборочного t-критерия.

```python
# задаём уровень значимости
alpha = 0.05 

# загружаем данные
data_chat_bot = [55, 53, 60, 49, 45, 57, 46, 53, 59, 53, 53, 55, 42, 41, 59, 43, 47, 60, 50, 57, 59, 56, 52, 48, 59, 53, 59, 50, 59, 59]
data_operator = [72, 80, 66, 72, 75, 71, 73, 71, 75, 68, 63, 68, 62, 65, 77, 66, 67, 62, 60, 74, 61, 67, 61, 63, 62, 79, 61, 63, 62, 63]

# проводим тест
_, p = stats.mannwhitneyu(data_chat_bot, data_operator)

print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```

Критерий Уилкоксона. Его используют, чтобы оценить различия распределений признака в двух зависимых группах. Это непараметрический аналог парного t-критерия.



```python
# задаём уровень значимости
alpha = 0.05 

# загружаем данные
data_chat_bot = [71, 97, 71, 97, 83, 90, 83, 94, 88, 76, 79, 99, 82, 85, 93, 78, 76, 87, 73, 72, 89, 89, 71, 86, 78, 93, 86, 95, 83, 73]
data_operator = [114, 112, 115, 124, 101, 108, 104, 109, 103, 115, 100, 129, 120, 129, 117, 125, 112, 105, 128, 107, 120, 108, 129, 100, 116, 105, 128, 128, 120, 106]

# проводим тест
_, p = stats.wilcoxon(data_chat_bot, data_operator, alternative = 'less')

print('p-value = {:.3f}'.format(p))

# интерпретируем результат
if p <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```
