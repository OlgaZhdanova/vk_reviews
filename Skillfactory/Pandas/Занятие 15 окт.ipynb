```python
# предобработка данных
# приводим столбцы к нужному типу
nobel['died'] = pd.to_datetime(nobel['died'])
```

```python
import numpy as np
# мы заменяем эти нули на nan'ы
nobel['died'].replace('0000-00-00', np.nan, inplace=True)
nobel.head()
```

```python
# приводим столбцы к нужному типу
nobel['died'] = pd.to_datetime(nobel['died'])
nobel.head(3)
```

```python
nobel['born'] = pd.to_datetime(nobel['born'])
```

```python
nobel['born'] = pd.to_datetime(nobel['born'], format='mixed')
```

```python
# отбираем строки где есть эти нули
incomplete_born_df = nobel[nobel['born'].str.contains('-00-00')]
incomplete_born_df.head(3)
```

```python
incomplete_born_df.shape[0]
```

```python
nobel['incomplete_born'] = np.where(nobel['born'].str.contains('-00-00'), nobel['born'], None)
nobel.head(3)
```

```python
nobel['born'] = nobel['born'].replace(to_replace='.*-00-00', value= np.nan, regex=True)
nobel.head(3)
```

```python
nobel['born'] = pd.to_datetime(nobel['born'], format='mixed')
nobel.info()
```

```python
nobel['incomplete_born']
```

```python
# вырезаем 4 первых значения из строки и приводим год к int16
nobel['incomplete_born'].str[:4].astype('Int16')
# <NA> - значение для пропущенныз данных в целочисленном типе
```

```python
# добавим колонку с именем born_year сразу после колонки born с помощью метода insert()
nobel.insert(nobel.columns.get_loc('born') + 1, 'born_year', nobel['incomplete_born'].str[:4].astype('Int16'))
nobel.head(3)
```

```python
# удаляем столбец incomplete_born за ненадобностью
nobel.drop(columns=['incomplete_born'], inplace=True)
nobel.head(1)
```

```python
# заполним пропуски (NA) в столбце born_year с помощью данных из столбца born
nobel['born_year'] = nobel['born_year'].fillna(nobel['born'].dt.year)
nobel.head(3)
```
