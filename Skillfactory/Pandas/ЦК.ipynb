# Работа с пропусками

```python
data.isnull().sum()
# видим что 1 пропуск в рейтинге и 34 - в сертификате
```

```python
# посмотрим какую долю составляют пропуски
data.isnull().mean().sort_values(ascending=False)
```

```python
# пропуски можно просто "выбросить"
# и посмотреть снова долю незаполненных ячеек
data.dropna(subset=['Certificate', 'Rating']).isnull().mean()
```

```python
# но отбрасывание пустых значений не всегда подходит, т к может сказаться на результатах исследования
# так что лучше пустые значения чем-то "заполнять"
# можно заполнить просто нулями, но если понадобиться исследовать рейтинг, это будет явным выбросом
data['Rating'].fillna(value=0)
```

```python
# можно заполнить медианным значением по столбцу
data['Rating'].fillna(value=data['Rating'].median())
```

```python
# а также предыдущим/следующим значением
data['Rating'].fillna(method='ffill')
```

# Работа с несколькими таблицами (join, merge, конкатенация)

```python
# для начала создадим три таблицы
raw_data={
    'user_id': ['1', '2', '3', '4', '5'], 
    'first_name': ['Jin', 'Alex', 'Shelly', 'Anna', 'Troy'],
    'last_name': ['Kazama', 'Lewis', 'Tenant', 'Ivanova', 'Brown']
}
df_a = pd.DataFrame(raw_data, columns=['user_id', 'first_name', 'last_name'])
df_a
```

```python
raw_data={
    'user_id': ['4', '5', '6', '7', '8'], 
    'first_name': ['Anna', 'Troy', 'Rio', 'Terry', 'Peter'],
    'last_name': ['Ivanova', 'Brown', 'Ferdinand', 'Parker', 'Parker']
}
df_b = pd.DataFrame(raw_data, columns=['user_id', 'first_name', 'last_name'])
df_b
```

```python
raw_data={
    'user_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10'], 
    'outer_id': ['223', '1134', '89', '671', '278', '17', '1931', '216', '81']
}
df_c = pd.DataFrame(raw_data, columns=['user_id', 'outer_id'])
df_c
```

```python
# конкатенация по строкам
df_new = pd.concat([df_a, df_b], ignore_index=True)
df_new
# заметим, что есть повторяющиеся строки
```

```python
# конкатенация по столбцам
pd.concat([df_a, df_b], axis=1)
```

```python
# мерджим две таблицы по столбцу user_id
pd.merge(df_new, df_c, on='user_id')
# заметим, что Rio Ferdinand у нас потерялся, потому что такого user_id (6) нет в df_c
```

```python
# merge можно выполнять по полю из левой и полю из правой таблиц. В примере названия колонок совпадают
pd.merge(df_new, df_c, left_on='user_id', right_on='user_id')
```

```python
# выписываются все пользователи
pd.merge(df_a, df_b, on='user_id', how='outer')
# значение NaN указывает, что в первоначальных таблицах не содержались данные
```

```python
# в таблицу попадают только те пользователи, ктр есть и в одной, и в другой таблицах
pd.merge(df_a, df_b, on='user_id', how='inner')
```

```python
# только те пользователи, что есть в левой
pd.merge(df_a, df_b, on='user_id', how='left')
```

```python
# только те пользователи, что есть в правой
pd.merge(df_a, df_b, on='user_id', how='right')
```
