```python
import json
from pprint import pprint
```

 Чтобы перевести данные из формата JSON в формат, который можно обрабатывать инструментами Python, необходимо выполнить процедуру, которая называется ДЕСЕРИАЛИЗАЦИЯ (декодирование данных). Обратный процесс, связанный с переводом структур данных Python в формат JSON, называется СЕРИАЛИЗАЦИЕЙ.
 Для десериализации пользуемся методом load()

```python
# открываем файл и связываем его с объектом f
with open('data/recipes.json') as f:
    # загружаем содержимое открытого файла в переменную recipes
    recipes = json.load(f)
```

```python
# подсчитаем кол-во уникальных национальных кухонь
unique_cuisine = set()
for recipe in recipes:
    for key in recipe:
        if key == 'cuisine':
            unique_cuisine.add(recipe[key])
print(unique_cuisine)
print(len(unique_cuisine))
```

```python
cuisine = {}
for recipe in recipes:
    for key in recipe:
        if key == 'cuisine':
            if recipe[key] in cuisine:
                cuisine[recipe[key]] += 1
            else:
                cuisine[recipe[key]] = 1
print(cuisine)
```

# XML

Аббревиатура XML расшифровывается как eXtensible Markup Language — расширяемый язык разметки. Он (язык) позволяет описывать документы, используя теги.

Данные в формате XML имеют древовидную структуру. Что такое дерево? Это структура, которая имеет узлы и связи между ними. Самый верхнеуровневый узел называется корнем, а всё, что находится в самом низу, называется листьями. 

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('data/menu.xml')
```

```python
# выведем на экран наименование всех блюд из меню, а также информацию о них
# во внешнем цикле перебираем потомков корневого узла
for dish in root:
    # во внутреннем цикле - потомков каждого блюда
    for param in dish:
        # выводим название блюда, название параметра и его значение
        print(dish.attrib['name'], param.tag, param.text)
    print()
```

```python
# преобразуем XML-файл в датафрейм
tree = ET.parse('data/menu.xml')
root = tree.getroot()
df_list = []
column_names = ['name', 'price', 'weight', 'class']
for dish in root:
    new_line = []
    new_line.append(dish.attrib['name'])
    for param in dish:
        new_line.append(param.text)
    df_list.append(new_line)
df = pd.DataFrame(
    data = df_list,
    columns = column_names
)
df
```

```python
# создаем корень дерева с нуля с помощью метода Element()
new_root = ET.Element('menu')
new_root
```

```python
# добавим двух потомков корневого узла через SubElement()
dish1 = ET.SubElement(new_root, 'dish', name='Кура')
dish2 = ET.SubElement(new_root, 'dish', name='Греча')
list(new_root)
```

```python
# добавим по три атрибута (потомка) к двум новым узлам
price1 = ET.SubElement(dish1, 'price').text = '40'
weight1 = ET.SubElement(dish1, 'weigth').text = '300'
xlass1 = ET.SubElement(dish1, 'class').text = 'Мясо'

price2 = ET.SubElement(dish2, 'price').text = '20'
weight2 = ET.SubElement(dish2, 'weigth').text = '200'
xlass2 = ET.SubElement(dish2, 'class').text = 'Крупа'

```

```python
# проверка
for dish in new_root:
    for param in dish:
        print(dish.attrib['name'], param.tag, param.text)
    print()    
```

```python
# сохраняем с помощью tostring
new_root_string = ET.tostring(new_root)
with open("new_menu.xml", "wb") as f:
    f.write(new_root_string)

# альтернатива - ET.ElementTree(new_root).write('new_menu_good.xml', encoding="utf-8")
```
